```python
# Graph of recent likes + polynomial regression through them 

# pull data from dataframe 
likes_list = df['Daily New Likes'].tolist()
engagement_list = df['Daily Page Engaged Users'].tolist()


x = np.arange(0, 164)
y = [0]
y1 = [0]
y2 = [0]
y3 = [0]

# arrange summed likes and summed engagements array 
for i in range(len(likes_list) - 1):
    try:
        temp = int(likes_list[i])
        y.append(temp + y[i])
    except:
        y.append(y[i])
    try:
        temp = int(engagement_list[i])
        y1.append(temp + y1[i])
    except:
        y1.append(y1[i])
        
# arrange regular likes and engagement array 
for i in range(len(likes_list) - 1):
    try:
        y2.append(int(engagement_list[i]))
    except:
        y2.append(0)
    try:
        y3.append(int(likes_list[i]))
    except:
        y3.append(0)
        
        
# numpy regression 
from sklearn.metrics import r2_score

model = np.poly1d(np.polyfit(x, y, 3))
line = np.linspace(0, 164, 100)
model1 = np.poly1d(np.polyfit(x, y1, 3))
line1 = np.linspace(0, 164, 100)
        
    
# plot likes
plt.plot(x, y3, color='#d62728')
plt.ylabel('likes on any given day since 12/1/2019')
plt.xlabel('days since 12/1/2019')
plt.title('likes per day')
plt.show()
    
    
# plot summed likes
plt.scatter(x, y, color='#d62728', marker='^')
plt.plot(line, model(line))
plt.ylabel('likes since 12/1/2019')
plt.xlabel('days since 12/1/2019')
plt.title('total likes')
plt.show()
print("R-squared value for the polynomial fit for likes is: " + str(r2_score(y, model(x))))


# plot engagement 
plt.plot(x, y2, color='g')
plt.ylabel('engagement on any given day since 12/1/2019')
plt.xlabel('days since 12/1/2019')
plt.title('engagement per day')
plt.show()


# plot summed engagement 
plt.scatter(x, y1, color='g', marker='^')
plt.plot(line1, model1(line1))
plt.ylabel('engagement since 12/1/2019')
plt.xlabel('days since 12/1/2019')
plt.title('engagement')
plt.show()
print("R-squared value for the polynomial fit for engagement is: " + str(r2_score(y1, model1(x))))

```

```python
# demographics bargraphs


# arrange data 
data = [] 
for col in cols:
    new_data_point = 0
    if "Lifetime Likes by Gender and Age" in col:
        _list = df[col].tolist()
        try:
            new_data_point = (int(_list[len(_list) - 1]) - int(_list[0]))
        except:
            pass
                
        # if page received negative likes from a certain demographic, 
        # display 0 
        if new_data_point >= 0:
            data.append(new_data_point)
        else:
            data.append(0)
        

N = 6
ind = np.arange(N) # the x locations for the groups
dataPoints = [data[0:6], data[6:12], data[12:18]]
width = 0.15
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(ind, dataPoints[0], width, color='r')
ax.bar(ind + width, dataPoints[1], width, color='b')
ax.bar(ind + 2*width, dataPoints[2], width, color='g')
ax.set_xticklabels(('why is this here', '18-24', '25-34', '35-44', '45-54', '55-64', '65+'))
ax.set_ylabel('Likers')
ax.set_title('Likes by group and gender (12/1/2019 - 5/12/2020)')
ax.set_yticks(np.arange(0, 20, 2))
ax.legend(labels=['Men', 'Women', 'Unknown'])
plt.show()
```

```python
# type of content which is most popular 


# arrange data 
topics = [ '-checkin-', '-coupon-', '-event-', '-fan-', '-mention-', 
          '-other-', '-page post-', '-question-', '-user post-']
data = [] 
for col in cols:
    
    new_data_point = 0
    
    if "Daily Talking About This By Story Type" in col:
        _list = df[col].tolist()
        
        for i in range(len(_list)):
            try:
                new_data_point += int(_list[i])
            except:
                pass
            
        data.append(new_data_point)

            
N1 = 9
ind1 = np.arange(N1) # the x locations for the groups
width = 0.2
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
ax.bar(ind1, data, width, color='r')
plt.xticks(ind1, (topics))
ax.set_ylabel('# of people')
ax.set_title('Daily People Talking About Types of Content (12/1/2019 - 5/12/2020)')
ax.set_yticks(np.arange(0, 1000, 100))
ax.legend(labels=['People'])
plt.show()
```

```python
# viral reach


# arrange data 

data = df["Daily Viral Reach Of Page Posts"].tolist()
data2 = []

for i in range(len(data)):
    try:
        data[i] = int(data[i])
    except:
        data[i] = 0
    try: 
        data2.append(data2[i - 1] + data[i])
    except:
        data2.append(data[0])

        
# plot viral reach per day
plt.plot(x, data, color='g')
plt.ylabel('reach of page posts since 12/1/2019')
plt.xlabel('days since 12/1/2019')
plt.title('viral reach per day')
plt.show() 

# plot viral reach cumulative 
plt.plot(x, data2, color='g')
plt.ylabel('reach of page posts since 12/1/2019')
plt.xlabel('days since 12/1/2019')
plt.title('viral reach cumulative')
plt.show() 


# 'velocity' (rate of change) of viral reach per day  

dx = np.diff(x)
dy = np.diff(data)
d = dy/dx

x1 = np.arange(0, 163)


plt.plot(x1, d, color='g')
plt.ylabel('change in reach of posts since 12/1/2019')
plt.xlabel('days since 12/1/2019')
plt.title('velocity of viral reach per day')
plt.show()            

```
